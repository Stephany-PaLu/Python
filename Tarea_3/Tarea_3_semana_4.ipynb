# 1. Acceder los datos de manera remota.

```python
datos_partidos= pd.read_csv('https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-04-10/partidos.txt',delimiter='\t')
datos_partidos
```

# 2. Crear dos gráficos en el que comparen a al menos dos países diferentes.

```python
equipos = datos_partidos[(datos_partidos.equipo_1 == 'Ecuador') | (datos_partidos.equipo_1 == 'Argentina') | (datos_partidos.equipo_1 == 'Brasil') | (datos_partidos.equipo_1 == 'Colombia') | (datos_partidos.equipo_1 == 'Uruguay')]
equipos
```

```python
prueba=["anfitrion", "estadio", "ciudad", "partido_orden", "fecha", "equipo_2", "equipo_2_final"]
prueba
```

```python
paises = equipos.equipo_1
paises
```

```python
equipos = equipos.drop(prueba, axis=1)
```

```python
equipos.head()
```

```python
equipos = equipos[['equipo_1', 'anio', 'equipo_1_final']]
equipos
```

```python
equipos_nuevo = equipos.pivot_table(index='anio', columns='equipo_1', values='equipo_1_final', aggfunc='sum')
equipos_nuevo
```

```python
equipos_nuevo.columns[equipos_nuevo.columns.str.contains('a')]
```

```python
equipos_nuevo.columns.name = None
```

```python
equipos_nuevo.columns[equipos_nuevo.columns.str.contains('a')]
```

```python
equipos_nuevo
```

```python
equipos_nuevo.index
```

```python
equipos_nuevo.reset_index(inplace = True)
```

```python
equipos_nuevo.rename(columns = {'index': 'anio'}, inplace=True)
```

# a. Gráfico 1 deberá ser un gráfico en el que los datos de los países comparados aparezcan en el mismo cuadro.

```python
fig = plt.figure()
fig.set_size_inches(14, 4, forward=True)
n = len(equipos_nuevo.index)
x = np.arange(n)
width = 0.35
plt.bar(x + width, equipos_nuevo.Brasil, width=width, label='Brasil', color='violet')
plt.bar(x, equipos_nuevo.Argentina, width=width, label='Argentina', color='orange')
plt.bar(x, equipos_nuevo.Uruguay, width=width, label='Uruguay', color='blue')
plt.bar(x, equipos_nuevo.Colombia, width=width, label='Colombia', color='turquoise')
plt.bar(x + width, equipos_nuevo.Ecuador, width=width, label='Ecuador', color='yellow')
plt.xticks(x + width, equipos_nuevo.anio)
plt.legend(loc='best')
#Ahora etiquetamos los ejes y le damos Titulo del grafico
plt.title('Número de Goles de los Países de América del Sur en Mundiales')
plt.xlabel('Año')
plt.ylabel('Número de Goles')
#Podemos guardar la figura en un archivo
plt.savefig('Goles_de_5_países_de_America_Sur.png')
```

# b. Gráfico 2 deberá ser un gráfico en el que los datos de cada país aparezca en un subgráfico diferente. Es decir, una figura tenga al menos dos recuadros (uno por cada país en la comparación).

```python
#Inicializamos una figura
fig = plt.figure()
fig.set_size_inches(15, 4, forward=True)
#Le damos un título
plt.title('Comparación de goles de cada país de América del Sur en los diferentes mundiales de fútbol')
#Agregamos un eje (2 filas, 2 columnas, primer eje)
eje = fig.add_subplot(1, 5, 1)
#Graficamos los datos de Ecuador en el primer eje
equipos_nuevo.plot(x = 'anio', y = 'Ecuador', ax = eje, ls = "-.", color="red")

#Agregamos un segundo eje (2 filas, 2 columnas, segundo eje)
eje2 = fig.add_subplot(1, 5, 2, sharey = eje)
#Graficamos los datos de Argentina en el segundo eje
equipos_nuevo.plot(x = 'anio', y = 'Argentina', ax = eje2, ls = "-.", color="orange")

#Agregamos un tercer eje (2 filas, 2 columnas, segundo eje)
eje3 = fig.add_subplot(1, 5, 3, sharey = eje)
#Graficamos los datos de Brasil en el segundo eje
equipos_nuevo.plot(x = 'anio', y = 'Brasil', ax = eje3, ls = "-.", color="fuchsia")

#Agregamos un cuarto eje (2 filas, 2 columnas, segundo eje)
eje4 = fig.add_subplot(1, 5, 4, sharey = eje)
#Graficamos los datos de Colombia en el segundo eje
equipos_nuevo.plot(x = 'anio', y = 'Colombia', ax = eje4, ls = "-.", color="blue")

#Agregamos un quinto eje (2 filas, 2 columnas, segundo eje)
eje5 = fig.add_subplot(1, 5, 5, sharey = eje)
#Graficamos los datos de Uruguay en el segundo eje
equipos_nuevo.plot(x = 'anio', y = 'Uruguay', ax = eje5, ls = "-.", color="green")

#Agregamos etiquetas a los ejes
eje.set_xlabel('Año')
eje2.set_xlabel('Año')
eje3.set_xlabel('Año')
eje4.set_xlabel('Año')
eje5.set_xlabel('Año')
eje.set_ylabel('Número de Goles')
#Guardamos la imagen
plt.savefig('subgraficos_paises_en_la_copa_mundial.png')
```
